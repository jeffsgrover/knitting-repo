```python
pattern_rgauge = 29
pattern_sgauge = 28

new_rgauge = 36
new_sgauge = 32

class square():
  '''
  inputs:
    - rcount: number of rows
    - scount: number of stitches

  row_gauge:    number of rows per 10 cm
  stitch_gauge: number of stitches per 10 cm

  row_length:    how long is row_count at row_gauge, in cm
  stitch_length: how long is stitch_count at stitch_gauge, in cm
  '''

  def __init__(self, rcount=1, scount=0):
    self.rcount  = rcount
    self.scount  = scount
    self.rgauge  = pattern_rgauge
    self.sgauge  = pattern_sgauge
    self.rlength = rcount/pattern_rgauge*10
    self.slength = scount/pattern_sgauge*10
    self.has_been_converted = False

    print('{} with {} rows and {} stitches'.format(type(self).__name__, self.rcount, self.scount))
    print('original gauge: {} rows and {} stitches per 10 cm'.format(self.rgauge, self.sgauge))

  def convert_gauge(self, new_rgauge=new_rgauge, new_sgauge=new_sgauge):
    '''
    convert from the pattern's gauge to your gauge
    '''
    def print_(self):
      print('new gauge: {} rows and {} stitches per 10 cm'.format(self.new_rgauge, self.new_sgauge))
      print('cast on {:.1f} stitches'.format(self.scount))
      print('knit {:.1f} rows'.format(self.rcount))
    if self.has_been_converted:
      print_(self)
    else:
      self.new_rgauge = new_rgauge
      self.new_sgauge = new_sgauge
      self.rcount = self.rcount * self.rgauge/new_rgauge
      self.scount = self.scount * self.sgauge/new_sgauge
      self.has_been_converted = True
      print_(self)
```

```python
import numpy as np

pattern_gauge2 = [29, 28]
new_gauge2     = [36, 32]

class square():
  '''
  inputs:
    - count:  [# of rows, # of stitches]
    - gauge:  [# of rows per 10 cm, # of sts per 10 cm]
    - length: [height of shape, width of shape]
  '''

  def __init__(self, count=None, gauge=None, length=None):
    self.has_been_converted = False

    vars = [count, gauge, length]
    for i, var in enumerate(vars):
      if var is not None and type(var)==list:
        vars[i] = np.array([float(j) for j in vars[i]])
    self.count, self.gauge, self.length = vars

    if gauge is None:
      self.gauge = np.array(pattern_gauge2)
    if count is None:
      self.count = self.gauge/10 * self.length
    if length is None:
      self.length = self.count / (self.gauge/10)

    if type(self).__name__=='square':
      print('{} with {:.2f} rows and {:.2f} stitches; {:.2f} cm high and {:.2f} cm wide'.format(
        type(self).__name__, *self.count, *self.length))
      print('original gauge: {:.2f} rows and {:.2f} stitches per 10 cm'.format(*self.gauge))

  def convert_gauge(self, new_gauge=None):
    '''
    convert from the pattern's gauge to your gauge
    '''
    def print_(self, gauge_to_print):
      gauge_to_print = np.array([float(i) for i in gauge_to_print])
      print('new gauge: {:.2f} rows and {:.2f} stitches per 10 cm'.format(*gauge_to_print))
      print('cast on {:.1f} stitches'.format(self.count[1]))
      print('knit {:.1f} rows'.format(self.count[0]))

    if type(new_gauge)==list:
      new_gauge = np.array(new_gauge)
    if new_gauge is None:
      new_gauge = new_gauge2

    if self.has_been_converted:
      print_(self, self.gauge)
    else:
      self.count = self.count * new_gauge/self.gauge
      self.gauge = new_gauge
      self.has_been_converted = True
      print_(self, new_gauge)
```

```python
def lcm_float(a, b, precision):
    a = round(a, precision)
    b = round(b, precision)
    return math.lcm(int(a*10**precision), int(b*10**precision))/10**precision
```
